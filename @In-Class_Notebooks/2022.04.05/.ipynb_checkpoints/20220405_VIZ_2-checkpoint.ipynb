```python
import pandas as pd
import matplotlib.pyplot as plt

gapminder = pd.read_csv('https://raw.githubusercontent.com/rogovich/Data/master/data/gapminderData.csv')
gapminder['continent'] = pd.Categorical(gapminder['continent'])

for year in gapminder['year'].unique():
    fig, ax = plt.subplots(figsize=(7, 4))
    gapminder_year = gapminder[gapminder['year'] == year]
    ax.scatter(gapminder_year['gdpPercap'], gapminder_year['lifeExp'],
              s=gapminder_year['pop'] / 1000000, 
              alpha=0.7,
              c=gapminder_year['continent'].cat.codes,
              cmap='rainbow_r') # color map
    ax.set_ylim(0, 100)
    ax.set_xlim(100, 100000)
    ax.set_xscale('log')
    ax.set_title(f'Зависимость средней продолжительности жизни\n от ВВП на душу населения. Год: {year}')
    ax.set_xlabel('ВВП на душу населения')
    ax.set_ylabel('Средняя продолжительность жизни')
    x_china = gapminder_year[gapminder_year['country'] == 'China']['gdpPercap']
    y_china = gapminder_year[gapminder_year['country'] == 'China']['lifeExp']
    ax.annotate('КНР', (x_china, y_china))
    fig.savefig(f'{year}.png', dpi=100)
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.scatter(crimes.murder,
           crimes.burglary, 
           s=crimes.population / 100000, 
           alpha=0.6, 
           c=crimes.motor_vehicle_theft)

burglary_max = crimes.burglary.max()
x_murder = crimes[crimes.burglary == burglary_max ].murder
y_burglary = crimes[crimes.burglary == burglary_max].burglary
state = crimes[crimes.burglary == burglary_max].state.item()
ax.annotate(state, (x_murder, y_burglary))
```

```python
fig, ax = plt.subplots(figsize=(14,8))
ax.scatter(crimes.murder, crimes.burglary, s=crimes.population/100000, alpha=0.6, c=crimes.motor_vehicle_theft)

for state in crimes.state:
    crimes_state = crimes[crimes.state == state]
    x_murder = crimes_state.murder
    y_burglary = crimes_state.burglary
    ax.annotate(state, (x_murder, y_burglary))
```
