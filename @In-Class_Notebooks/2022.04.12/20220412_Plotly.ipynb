# Plotly

```python
import plotly.graph_objects as go
import pandas as pd
```

```python
crimes = pd.read_csv('https://raw.githubusercontent.com/rogovich/Data/master/data/crimeRatesByState2005.tsv', sep = '\t')
```

```python
crimes.head()
```

```python
trace0 = go.Scatter(x = crimes['murder'],
                    y = crimes['burglary'],
                    mode = 'markers', # график рассеяния
                    text = crimes['state'],
                    hovertemplate = '<b>%{text}</b><br>' +
                                    '<b>Murder Rate:</b> %{x}<br>' + 
                                    '<b>Burglary Rate:</b> %{y}<br>' +
                                    '<b>Motor Vehicle Theft Rate:</b> %{marker.color}',
                    marker = dict(size = crimes['population'] / 500000,
                                  opacity = 0.7,
                                  color = crimes['motor_vehicle_theft'],
                                  colorscale = 'Viridis',
                                  showscale = True)
                   ) 

layout = go.Layout(
    title = 'Crime in the USA',
    xaxis = dict(
        title = 'Murder rate (number per 100,000 population)',
        gridwidth = 1,
    ),
    yaxis =dict(
        title = 'Burglary rate (number per 100,000 population)',
        gridwidth = 1,
    ))

fig = go.Figure(data=trace0, layout=layout)
fig.show()
```
